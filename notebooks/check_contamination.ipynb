```python
import os
import tqdm
```

```python
def read_txt(speech_path):
    speech_paths = []
    with open(speech_path, 'r') as file:
        lines = file.readlines()
        for line in lines:
            speech_paths.append(line.rstrip()) 
    return speech_paths
```

```python
train_speech = read_txt('/home/ubuntu/Data/DFN/textfiles/training_set.txt')
```

```python
val_speech = read_txt('/home/ubuntu/Data/DFN/textfiles/validation_set.txt')
```

```python
test_speech = read_txt('/home/ubuntu/Data/DFN/textfiles/test_set.txt')
```

```python
for i in tqdm.tqdm(val_speech):
    if i in train_speech:
        print(i)
    if i in test_speech:
        print(i)
```

```python
for i in tqdm.tqdm(test_speech):
    if i in train_speech:
        print(i)
    if i in val_speech:
        print(i)
```

```python
train_noise = read_txt('/home/ubuntu/Data/DFN/textfiles/training_set_noise.txt')
val_noise = read_txt('/home/ubuntu/Data/DFN/textfiles/validation_set_noise.txt')
test_noise = read_txt('/home/ubuntu/Data/DFN/textfiles/test_set_noise.txt')
```
